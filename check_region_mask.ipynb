```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import nrrd
import os
```

```python
SERVER = 9
DATASET_NAME = "20220406_111526"
```

```python
filename = "/nfs/data9/jen/data/20220406_111526/NMF.h5"

with h5py.File(filename, "r") as f:
    ds_arr = f['stack_mean'][()]  # returns as a numpy array
    
    roi_centroid_x_all = f['stack_mean'][()]  # returns as a numpy array

    centroid_x_all = f['centroid_x_all'][()]  # returns as a numpy array
    centroid_y_all = f['centroid_y_all'][()]  # returns as a numpy array
    z_all = f['z_all'][()]  # returns as a numpy array
    
    print(f.keys())
```

```python
ds_arr.shape
```

```python
filename = "/nfs/data9/chuyu/data/20220406_111526/region_roi_bool.h5"
with h5py.File(filename, "r") as f:
    region_roi_bool = f['region_roi_bool'][()]  # returns as a numpy array
    region_names = f['region_names'][()]  # returns as a numpy array
    
```

```python
sum(region_names == b'Telencephalon - Pallium')
```

```python
pallium_rois = region_roi_bool[region_names == b'Telencephalon - Pallium'][0]
```

```python
plt.scatter(centroid_x_all, centroid_y_all, s =  1)
plt.scatter(centroid_x_all[pallium_rois==1], centroid_y_all[pallium_rois==1], s =  1)
```

```python
plt.scatter(centroid_x_all, z_all, s =  1)
plt.scatter(centroid_x_all[pallium_rois==1], z_all[pallium_rois==1], s =  1)
```

```python
ds_arr.shape
```

```python
which_z =  20
plt.imshow(ds_arr[which_z])
plt.scatter(centroid_x_all[(z_all==which_z)*(pallium_rois==1)], centroid_y_all[(z_all==which_z)*(pallium_rois==1)], s =  1, color="r")
```

```python
ds_arr.shape
```

```python
pallium_roi_mask = np.zeros(((ds_arr).shape[2], (ds_arr).shape[1], (ds_arr).shape[0]))
pallium_roi_mask[centroid_x_all[(pallium_rois==1)], centroid_y_all[(pallium_rois==1)], z_all[(pallium_rois==1)]] = 1
```

```python
pallium_roi_mask.shape
```

```python
data = pallium_roi_mask
data_int8 = (255*data/np.max(data)).astype(np.int8);
```

```python
nrrd.write(os.path.join(f"/nfs/data{SERVER}/chuyu/data/{DATASET_NAME}", f"{DATASET_NAME}_pallium_mask.nrrd"), data_int8)
```
